```python
from scapy.all import sniff, IP, TCP, UDP
import pandas as pd
from sklearn.ensemble import IsolationForest
import numpy as np
import time
import matplotlib.pyplot as plt

# Global variables
packets_data = []  # To store packet data
anomalies_detected = []  # To store detected anomalies

# Function to extract features from packets
def extract_features(packet):
    features = {}
    if IP in packet:
        features['src_ip'] = packet[IP].src
        features['dst_ip'] = packet[IP].dst
        features['protocol'] = packet[IP].proto
        features['length'] = len(packet)
        if TCP in packet:
            features['src_port'] = packet[TCP].sport
            features['dst_port'] = packet[TCP].dport
        elif UDP in packet:
            features['src_port'] = packet[UDP].sport
            features['dst_port'] = packet[UDP].dport
    return features

# Function to analyze packets in real-time
def packet_callback(packet):
    global packets_data
    features = extract_features(packet)
    if features:  # Only append if features are extracted
        packets_data.append(features)
        print(f"Captured packet: {features}")

# Function to detect anomalies using Isolation Forest
def detect_anomalies(packets_data):
    if len(packets_data) < 1:  # Wait for at least 1 packets to start detection
        return []

    # Convert packet data to DataFrame
    df = pd.DataFrame(packets_data)
    df = df.fillna(0)  # Fill missing values with 0

    # Feature matrix (use packet length and ports as features)
    X = df[['length', 'src_port', 'dst_port']].values

    # Train Isolation Forest model
    clf = IsolationForest(contamination=0.05)  # 5% expected anomalies
    df['anomaly'] = clf.fit_predict(X)
    df['anomaly'] = df['anomaly'].apply(lambda x: 'Anomaly' if x == -1 else 'Normal')

    # Filter anomalies
    anomalies = df[df['anomaly'] == 'Anomaly']
    return anomalies.to_dict('records')

# Function to generate alerts for anomalies
def alert_anomalies(anomalies):
    for anomaly in anomalies:
        print(f"🚨 ALERT: Anomaly detected! Source IP: {anomaly['src_ip']}, Destination IP: {anomaly['dst_ip']}, "
              f"Length: {anomaly['length']}, Protocol: {anomaly['protocol']}")

# Function to visualize network traffic with correct anomaly marking
def visualize_traffic(packets_data, anomalies_detected):
    if not packets_data:
        print("No packet data available for visualization.")
        return

    df = pd.DataFrame(packets_data)
    df.index = pd.to_datetime(df.index, unit='s')  # Ensure proper timestamp indexing
    df['timestamp'] = df.index

    # Convert anomalies into a DataFrame
    anomalies_df = pd.DataFrame(anomalies_detected)

    plt.figure(figsize=(12, 6))

    # Plot all packet lengths
    plt.plot(df['timestamp'], df['length'], label='Packet Length', color='blue')

    # Plot anomalies as red dots (if anomalies exist)
    if not anomalies_df.empty:
        anomalies_df.index = pd.to_datetime(anomalies_df.index, unit='s')
        anomalies_df['timestamp'] = anomalies_df.index
        plt.scatter(anomalies_df['timestamp'], anomalies_df['length'], color='red', label='Anomaly', marker='o')

    plt.xlabel('Time')
    plt.ylabel('Packet Length')
    plt.title('Network Traffic Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()

# Main function to run the analysis
def main():
    global packets_data, anomalies_detected

    print("Starting real-time network traffic analysis...")
    print("Press Ctrl+m i to stop.")

    # Start sniffing packets in real-time
    sniff(prn=packet_callback, store=False)

    # Periodically analyze packets and detect anomalies
    try:
        while True:
            time.sleep(5)  # Analyze every 10 seconds
            anomalies = detect_anomalies(packets_data)
            if anomalies:
                anomalies_detected.extend(anomalies)
                alert_anomalies(anomalies)
                # Pass the anomalies_detected argument to the function call
                visualize_traffic(packets_data, anomalies_detected)
    except KeyboardInterrupt:
        print("Stopping network traffic analysis.")

if __name__ == "__main__":
    main()
```
